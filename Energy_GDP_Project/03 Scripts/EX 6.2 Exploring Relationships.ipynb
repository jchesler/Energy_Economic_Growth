```python
#create a subset for analysis
df_subset = df[['ea_%_pop', '%_energy_nuclear', '%_energy_renewables', '%_energy_fossil_fuels', 'gdp']]
```

```python
#check output
df_subset.head()
```

```python
#create a matric heatmap using matplot
f = plt.figure(figsize=(8, 8))
plt.matshow(df_subset.corr(), fignum=f.number)
plt.xticks(range(df_subset.shape[1]), df_subset.columns, fontsize=14, rotation = 45)
plt.yticks(range(df_subset.shape[1]), df_subset.columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=14)
```

```python
#create a heatmap using seaborn

f,ax = plt.subplots(figsize=(8,8))

corr = sns.heatmap(df_subset.corr(), annot = True, cmap='viridis', fmt='.2f', linewidths =.5, ax = ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
plt.title('Correlation Matrix', fontsize=14)
```

For these coefficients in each sqaure represents the correlation between the two varialbes. the large the coefficient, the stronger the positive correlation between the two variables. Some values are negative, which represent an inverse correlation. 

In this chart the strongest correlation is between percent of the population that has access to energy and the gdp per capita. In other words, as the percent of people have access to electricity increases, so does the gdp per capita and vice versa.
